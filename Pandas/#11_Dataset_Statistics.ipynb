# DataStatistikasi ( sum , min , max , mean )
## Boshqa funksiyalar

`count` - NaN bo'lmagan qiymatlar soni

`median` - qiymatlar medianini hisbolash

`mad` - mean absolute deviation

`prod` - barcha qiymatlar ko'paytmasi

`var` - variance (dispersiya)

`std` - standart deviation (standart chetlanis)

`skew` - skewness

`kurt` - kurtosis

`cumsum` - cumulative sum

`cummin` va `cummax`- cumulative min and max

`cumprod` - cumulative product

`diff` - arifmetik farq (difference)

`pct_change` - percent changes

# Karrelyasiya yani bir biriga bogliqligi ni takshirish
- ikkita DataFrame ni bir biri bilan bog'liqligi ni takshiradi yoki bitta datafrane ustunlar buyicha ham solishtrsa buladi
- `df.corrwith(df['price'])` # faqat bitta ustun buyicha solishtrish
- `df.corr() `# umumiy jadvalni solishtrish
- `df['year'].corr(df['price'])` >> ikta ustunni bir bn bogliqligi

```python
df
```

```python
df['year'].corr(df['price']) # yani chiqarilgan yili uning narxiga 52 % bogliq
# agar natija mustat bulsa usish yani yangi yildagi qimmat ,
# agar natija manfiy bulsa kamayish yani eski mashina qimmat bulishi malum bir % ga bogliq
```

```python
df.corr() # umumiy jadvalni solishtrish
```
