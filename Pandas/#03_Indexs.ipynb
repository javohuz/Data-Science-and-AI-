# Bazi Index metodlar va parametrlari
- `newdf.isin(df3.index)` >> Indeks boshqa indeks tarkbidia ekanini tekshirish
- `df3.index.difference(newdf)` >> Indeksning boshqa indeksdan farqini tekshirish
- `df3.index.union(newdf)` >> Ikki indeks to'plami
-  `.delate(), .drop()` >> o'chirish
- `df2.index.delete(0)` , `df2.index.drop(1)`
- `.inser() ,  df3.index.insert(2,'avaz')` >> Ko'rsatilgan index urniga malumot qushadi
