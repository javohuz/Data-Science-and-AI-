# Dataset DataFrame Filtrlash

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv')
df.head()
```

```python
models = df['model'].unique() # takrorlanuvchilanri olib tashlash
models.sort()
models
```

```python
years = df['year'].unique()
years.sort()
years
```

```python
df['year'].value_counts() #  ustunlardagi bir xil qiymatlilar nechta ekanligi yoki shu misolda malum bir yilda nechta mashina borligi
```

```python
df['model'].value_counts()
```

```python
mask = df['model'].isin([' A Class' ,' B Class' ])
df[mask]
```

```python
df[df['year'].isin([2019,2020])] # faqat 2019 hamda 2020 yildagi mashinalar
```

```python
df1 = df[ (df['price']>=20000) & (df.year == 2020) ]
df1.sort_values(by='price')
```
