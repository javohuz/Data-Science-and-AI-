- Data frame ko'p ulchamli malumotlar tuzulmasi asoas jadval kurinishida foydalaniladi
-  `df = DataFrame(date)` >> malumotlarni dataframe ga utqazish (python dict) jadval kurishinida utadi
- `df.head(n)` >> Dastlabgi n ta elementini kursatadi agar bulmasa 5 ta
- `df['Yil']` or `df.Yil` >> DataFramening istalgan ustunini Series ko'rinishida ajratib olishimiz mumkin.
- Lug'atdan DataFrame yaratishda har bir qator indekslanadi (0,1,2...), ustunlar ketma-ketligi esa lug'atdagi ketma-ketlikka bog'liq bo'ladi. Ustunlar ketma-ketligini berishni istasak `DataFrame` yaratishda `columns` parametrini ham ko'rsatib ketishimiz mumkin:

- Agar `columns ` ro'yxatida lug'atda mavjud bo'lmagan ustin bo'lsa, shu ustunga **bo'sh qiymatlar** qo'yib ketiladi.
- `'Name' in df.columns` or `number in df.index` index yoki columns ichida name yoki number elementi bor yuqligini tekshirish

```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
data = {'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
       'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
       "Temp":[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53]}
df = DataFrame(data)
df.head()
df['Yil']
```

```python
df = DataFrame(data , columns=['Yil','Temp', 'Aholi soni'])
df
```

```python
df = DataFrame(data , columns=['Yil','Temp', 'Aholi soni' , 'Jami'])
df
```

```python
df = DataFrame(data , columns=['Yil','Temp', 'Aholi soni' , 'Jami'] , index= [1,2,3,4,5,6,7,8])
df
```

```python
df.columns
```

`Series` yordamida biror ustunning ma'lum bir qatorlariga o'zgartirish kiritish mumkin. Bunda qatorlarning indeksi berilishi **shart**.

```python
df.Jami = pd.Series([1500,233,1411] , index=[1,2,6])
df
```

Malumotlarni uzgartish yoki malumot qushish

```python
df.Jami = 1500
df.head()

vol = [1500,1200,1222,4544,5444,5555,333,3232]
df.Jami = vol
df
```

Agar qiymat berishda mavjud bo'lmagan ustun nomi yozilsa, `DataFramega` yangi ustun qo'shiladi (bu faqat lug'at ko'rinishda yozishga xos).

Quyidagi misolda biz avval yangi `1500+` degan ustun yaratdik keyin esa bu ustun qiymatlarini `df.YAIM>1500` sharti yordamida to'ldridik ya'ni yalpi ichki mahsluot 1500 dan katta bo'lsa `True` aks holda `False`.


```python
df['1100+'] = df.Jami>1100
df
```
