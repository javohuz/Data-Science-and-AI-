# Qayta indexing

**DIQQAT! DataFrame metodlati asl DataFrameni o'zgartirmaydi, balki yangi (o'zgartirilgan) DataFrame qaytaradi.**
- `.reindex()` >> index urnini almanshtrish
- `df.head()` >> boshidagi 5 ta emement
- `df.sample(n)` >> ixriyoriy bitta or n ta elemantni qaytadi
- `axis=0 , axis=1 ` axis manosi o'q deani axis= 0 index , axis=1 columns

- `df.drop(10)`  # 10 yildagi yiymatlar tashlab yoborildi
- `df.drop([10,20,15])` # list ichidagi indexs qiylati tashlab yoborildi <br>
**yuqoridagi uzgartrish asosiy dfni uzgartirmaydi uzgartish uchun inplase=True yoki df = ... deb asosini uzni uzgartish ham mumkun **

- `df.drop(10, inplace=True)` >> asosiy df ham uzgaradi bu bn ehtiyo bulish kerak
- `df.drop('Yil' , axis=1) ` >> Columns dagi elementlani uchirsh binda ham inplace tepadagidek ishleydi
- `df.drop('Yil' , axis='columns') ` >> ikalasi bir hil natija beradi

````python
import numpy as np
import pandas as pd
````

````python
data = {'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
       'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
       'Temp':[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53],
       'Zichlik':[79.77, 78.68, 77.53, 76.34, 75.13, 73.91, 72.71, 67.03]}

df = pd.DataFrame(data)
df
````

````python
df.index = [21, 20, 19, 18, 17, 16, 15, 10]
df
````

````python
df.reindex([10,15,21]) # kesib olish yangi dataframe qilib olib ishlatish mumkun
````

````python
s = [21, 20, 19, 18, 17, 16, 15, 10]
s.sort()
s
````

````python
df.reindex(s) # asoysiy df uzgarmaydi
````
