# Indekslash, qator (ustun) tanlash, filtrlash Series buyicha
DataFramening har bir qator va ustuni Series malumotlar tuzilmasi buladi
 - `data = df['Yil'] ` >>  malumotni Series kabi kuchirib olish
 - `data = df[['Yil']]` >>  malumotni DataFrame kabi kuchirib olish ,ikkinchi [ ]  ichida boshqa maulotlar ham bulishi mumkun
 - `data[6]` >> series ichidagi qiymatlar oddiy numpaydi son qiymatlar buladi amalla bajarish mumkun
 - `data[[6]] ` >>  ikkita qavs series buladi chalkashtirma , ikkinchi qavs ichida koplab elemenlar bu shunchaki ichidan ekrakli malumotlar lis kurinishid olishga uhshaydi
 - `obj['a':'d']` >>  agar indexsi harf buyicha sorting qilingan bulsa quyidagi yul bilan kerakli maumotlarni olish mumkun

 -  Filtrlash `obj[obj>4]` # uzizga kerak ligicha filtrlashimiz mumkun
 - `obj.to_numpy()` >> series mal tuzilmasidan numpy arrayga utish


```python
import numpy as np
import pandas as pd
data = {'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
       'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
       'Temp':[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53],
       'Zichlik':[79.77, 78.68, 77.53, 76.34, 75.13, 73.91, 72.71, 67.03]}

df = pd.DataFrame(data)
df
```

```python
data = df['Yil'] # malumotni Series kabi kuchirib olish
print(data , type(data))

```

```python
data = df[['Yil']] #  malumotni DataFrame kabi kuchirib olish ,
# ikkinchi []  ichida boshqa maulotlar ham bulishi mumkun
print(data , type(data))
```

```python
data[6]  # series ichidagi qiymatlar oddiy numpaydi son qiymatlar buladi amalla bajarish mumkun
type(data[6])
```

```python
data[[6]] # ikkita qavs series buladi chalkashtirma ,
#  ikkinchi qavs ichida koplab elemenlar bu shunchaki ichidan ekrakli malumotlar lis kurinishid olishga uhshaydi
```

```python
data[0:6] # ichidagi bir nechta elementlani series qilib olish
```

```python
obj = pd.Series(np.arange(8.), index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])
obj
```

```python
obj['a':'d'] # agar indexsi harf buyicha sorting qilingan bulsa quyidagi yul bilan ham kerakli maumotlarni olish mumkun
# va buning ustida turli amallar bajarish mumkun
obj[1:3]=3
obj
```

Filtralsh

```python
obj[obj>4] # uzizga kerak ligicha filtrlashimiz mumkun
```

```python
obj.to_numpy() # series mal tuzilmasidan numpy arrayga utish
```

# DataFrame
Malutolarni Series yoki DataFrame kurishiga almashtri ishlashdan maqsad seriesdga ham DF ni ham uziga hos methodlari bor ular bir biriga tushmay qolish mumukun
- ` df[['Uzbekistan']].to_numpy()`  >> DataFrame malumotlar tuzilamasidan numpyfaga arrayga utish
- `df2 = df.transpose(copy=True)` >> qator va ustunlarni almashtrish


```python
uz = {2020:33.5, 2019:32.9, 2018:32.5, 2017:31.9, 2016:31.4, 2015:30.9}
kz = {2020:18.8, 2019:18.5, 2018:18.3, 2017:18.1, 2016:17.8, 2015:17.6}
tj = {2020:9.5, 2019:9.3, 2018:9.1, 2017:8.9, 2016:8.7, 2015:8.5}
kg = {2020:6.5, 2019:6.4, 2018:6.3, 2017:6.2, 2016:6.1, 2015:6.0}
tr = {2020:6.0, 2019:5.9, 2018:5.9, 2017:5.8, 2016:5.6, 2015:5.6}

df = pd.DataFrame(
    {'Uzbekistan':uz,
     'Kazakhstan':kz,
     'Tajikistan':tj,
     'Kyrgizstan':kg,
     'Turkmenistan':tr
    }
)
```

```python
df[['Uzbekistan' , 'Turkmenistan']].to_numpy()
```

```python
df[['Uzbekistan' , 'Turkmenistan']]
```

```python
df[2:] # ichidan kerakli malumotrni olib turish
```

Filtrlash

```python
df>10 # aholis 10 mlndan kop mamlakatlar
```

```python
df[['Uzbekistan']][df['Uzbekistan']>32] # filtr uz aholis 32 mlndan katta bulgan yillar
```

```python
df2 = df.transpose(copy=True) # qator va ustunlarni almashtrish
df2
```

```python
df2[df2[2015]>6] # 2015 yilda aholisi 15 mln dan kop bulgan mamlakatlarni filtr qilish
```
