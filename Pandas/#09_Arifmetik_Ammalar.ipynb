# Arifetik ammallar
- series malumotlar tuzilmasida amllar index lari buyicha buladi yani mos indexlaridagi qiymatlar qushiladi yoki ....
- DataFrame da ham mos ustun (columns) va qatorlar (index) buyicha amllar bajaradi agar mos indexda mos qator bulmas ham NaN buladi
<hr>
- quyidagi ammalr ikkalasi uchun ham urinli
- `add`, `radd` - Qo'shish (+)
- `sub`, `rsub` - Ayirish (-)
- `div`, `rdiv` - Bo'lish (/)
- `floordiv`, `rfloordiv` - Bo'lish va butun qismini olis (//)
- `mul`, `rmul` - Ko'paytirish (*)
- `pow`, `rpow` - Eksponenta (\*\*)
- `fill_value=0` funksiyasi agar mos indexda mos bulmagan columns bulsa bittasini urniga 0 quyib amallni bajaradi

Yuqoridagi amallarning barchasida `fill_value` parametri mavjud.

```python
import numpy as np
import pandas as pd
```

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 'e'])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])
```

```python
s1+s2 # elementlar mos indexlar buyicha qushildi bir xil indexlar bulmasa NaN qiymat oladi
```

```python
s1.add(s2) # s1+s2 qushish
s1.sub(s2) # s1-s2 ayirish
s1.mul(s2)
```

DataFrame

```python
df1 = pd.DataFrame(np.arange(9.).reshape(3,3), columns=list('bcd'), index=['Olma','Anor','Uzum'])
df2 = pd.DataFrame(np.arange(12.).reshape(4,3), columns=list('abc'), index=['Olma','Anor','Qovun','Anjir'])
```

```python
df1
```

```python
df2
```

```python
df1+df2 # bu yerda ham mos ustun va qatorlar buyicha amllar bajaradi agar mos indexda mos qator bulmas ham NaN buladi ,
```

```python
df1.add(df2, fill_value=0)
```

Series va DataFrame urtasida ammalar

```python
df = pd.DataFrame(np.arange(1,13.).reshape(3,4), index=list('abc'), columns=['Olma','Anor','Qovun','Anjir'])
obj = pd.Series([1,1,2,2], index=df.columns)
print(df , obj )
```

```python
df + obj # bu yerda mos columns lardagi yani qatorlar qiymarlarga seriesning indexlaridagi mos qiymatlariga qushib chiqiladi
# va istalgan yuqiradi istallgan ammallarni bajarish mumkun faqat ( fill_value ishlamaydi !!!)
```

```python
df.mul(obj)
```
