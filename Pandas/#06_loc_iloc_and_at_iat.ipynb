## `.loc` va `.iloc`
- `.loc` >> loacation suzidan olingan yani manzil yordamida qator va ustunlani ularning nomi bo'yicha tanlash mumkin.
ELementlarni tanlab olish

- `.iloc` >> locga indexlar bilan murojat qilish ucun

```python
import numpy as np
import pandas as pd
```

```python
uz = {2020:33.5, 2019:32.9, 2018:32.5, 2017:31.9, 2016:31.4, 2015:30.9}
kz = {2020:18.8, 2019:18.5, 2018:18.3, 2017:18.1, 2016:17.8, 2015:17.6}
tj = {2020:9.5, 2019:9.3, 2018:9.1, 2017:8.9, 2016:8.7, 2015:8.5}
kg = {2020:6.5, 2019:6.4, 2018:6.3, 2017:6.2, 2016:6.1, 2015:6.0}
tr = {2020:6.0, 2019:5.9, 2018:5.9, 2017:5.8, 2016:5.6, 2015:5.6}

df = pd.DataFrame(
    {'Uzbekistan':uz,
     'Kazakhstan':kz,
     'Tajikistan':tj,
     'Kyrgizstan':kg,
     'Turkmenistan':tr
    }
)
df
```

```python
df.loc[[2020]] # 2020 yildagi malumoltarli olish
```

```python
df.loc[[2020] , ['Uzbekistan' , 'Tajikistan' ]] # 2020 yildagi uzb va taj dagi maulumotlarni olish
```

```python
df.iloc[[1,2,3]] # faqat bunga index bn murojat qilsih kerak
```

```python
df.iloc[1:3 , 2:4] # kesib olish yuli bilan
```

## Sonli indekslar

`pandas` obyektlariga indeks orqali murojaat qilish biroz mujmal tuyulishi mumkin. Bunga sabab biz o'rganib qolgan list va tuple ma'lumot turlari bilan ishlashdagi farq.<br>
Misol uchun, Pythonda ro'yxatning oxirgi elementini ko'rish uchun -1 indeksi orqali murojaat qilishimiz mumkin:

```python
nums1 = list(range(0,11))
nums1
```

```python
nums1[-1] # nums oxirgi elementi
```

Yuqoridagi usulni sonli indeksga ega `Series` ma'lumot tuzilmasiga qo'llab ko'ramiz:

```python
nums = pd.Series(np.arange(10.))
nums
```

```python
nums[-1]
```

Ko'rib turganingizdek bu usul raqamli indekslar bilan ishlamas ekan. Agar indesklar raqamli emas, matn bo'lgab holatda ishlaydi yoki <br>
Muammoning yechimi doimo `.loc` yoki `.iloc` parametrlaridan foydalanish

```python
nums.iloc[-1]
```

# `.at , .iat ` >> joyga ishora qiladi , bitta qiymatni ajratish

```python
df.at[2015 , 'Uzbekistan']
```

```python
df.at[2016 , 'Uzbekistan'] - df.at[2015, 'Uzbekistan'] # bitta qiymatni ajratib olib ustida ammallar bajarish mumkun
```

.iat => index bilan murojat qilish uchun

```python
df.iat[1,4] # va hokozi
```
